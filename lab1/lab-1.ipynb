## Задание 3

#### Воспроизведите функцию rectangle из OpenCV с нуля. Разрешено использовать только NumPy для разработки из сторонних библиотек (не обязательно, чтобы параметры были те же самые, как в OpenCV, например thickness. Главное, чтобы прямоугольник было видно на изображении). Продемонстрируйте работу вашей функции на тестовом изображении

```python
def my_rectangle(pic,cor1, cor2, color=(0,0,255), thickness=1):
    x1, y1 = cor1
    x2, y2 = cor2

    x_min, x_max = min(x1,x2), max(x1,x2)
    y_min, y_max = min(y1,y2), max(y1,y2)

    th_up= thickness//2
    th_down= thickness- th_up

    height, width= pic.shape[:2]

    y_start= max(0, y_min- th_up)
    y_end= min(height,y_min+ th_down)
    x_start= max(0, x_min- th_up)
    x_end= min(width, x_max + th_down)
    pic[y_start:y_end,x_start:x_end] = color

    y_start = max(0, y_max - th_up)
    y_end= min(height, y_max + th_down)
    pic[y_start:y_end,x_start:x_end] = color

    x_start =max(0, x_min - th_up)
    x_end= min(width, x_min + th_down)
    y_start= max(0, y_min - th_up)
    y_end= min(height, y_max + th_down)
    pic[y_start:y_end,x_start:x_end] = color

    x_start = max(0, x_max - th_up)
    x_end= min(width, x_max + th_down)
    pic[y_start:y_end,x_start:x_end] = color

    return pic
```

```python
img_copy2 = img.copy()
```

```python
img_copy2=my_rectangle(img_copy2,(110,492), (2475,1960), (0, 255, 0),5)
plt.imshow(img_copy2)
```
